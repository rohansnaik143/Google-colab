```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# Specify the URL of the webpage
url = "https://en.wikipedia.org/wiki/List_of_best-selling_automobiles"

# Send an HTTP request to the webpage
response = requests.get(url)

# Parse the HTML content of the webpage
soup = BeautifulSoup(response.content, 'html.parser')

# Find the table containing the data
table = soup.find('table', {'class': 'wikitable'})

# Extract the header of the table
header = [th.text.strip() for th in table.find_all('th')][:5]

# Extract the rows of the table
rows = []
for tr in table.find_all('tr'):
    cols = [td.text.strip() for td in tr.find_all('td')][:5]
    if cols:
        rows.append(cols)

# Convert the extracted data into a pandas DataFrame
df = pd.DataFrame(rows, columns=header)

# Print the scraped data
print(df)
```

```python
from bs4 import BeautifulSoup
import requests

Link = "https://en.wikipedia.org/wiki/List_of_best-selling_automobiles"
Link_text = requests.get(Link).text
soup = BeautifulSoup(Link_text, 'html.parser')

# Find the table containing the data
table = soup.find('table', {'class': 'wikitable'})

# Extract the header of the table
header = [th.text.strip() for th in table.find_all('th')]

# Extract the rows of the table
rows = []
for tr in table.find_all('tr'):
    cols = [td.text.strip() for td in tr.find_all('td')]
    if cols:
        rows.append(cols)

# Convert the extracted data into a pandas DataFrame
import pandas as pd
df = pd.DataFrame(rows, columns=header[:5])

# Print the scraped data
print(df)
```

```python
from bs4 import BeautifulSoup
import requests
from PIL import Image

Link = "https://en.wikipedia.org/wiki/List_of_best-selling_automobiles"
Link_text = requests.get(Link).text
soup = BeautifulSoup(Link_text, 'html.parser')

# Find all image elements in the webpage
images = soup.find_all('img')

# Extract the image source URLs
img_urls = [img['src'] for img in images]

# Download and display each image
for url in img_urls:
    if url.startswith("http"):
        image_content = requests.get(url).content
        with open("image.jpg", "wb") as f:
            f.write(image_content)
        img = Image.open("image.jpg")
        img.show()
```

```python
from bs4 import BeautifulSoup
import requests
from PIL import Image
import os

Link = "https://en.wikipedia.org/wiki/List_of_best-selling_automobiles"
Link_text = requests.get(Link).text
soup = BeautifulSoup(Link_text, 'html.parser')

# Find all image elements in the webpage
images = soup.find_all('img')

# Extract the image source URLs
img_urls = [img['src'] for img in images]

# Download and display each image
for i, url in enumerate(img_urls):
    if url.startswith("http"):
        response = requests.get(url)
        filename = f"image_{i}.jpg"
        with open(filename, "wb") as f:
            f.write(response.content)
        img = Image.open(filename)
        img.show()
        print(f"Downloaded and displayed {filename}")
```

```python
from bs4 import BeautifulSoup
import requests
import os

# Specify the URL of the webpage
url = "https://en.wikipedia.org/wiki/List_of_best-selling_automobiles"

# Send an HTTP request to the webpage
response = requests.get(url)

# Parse the HTML content of the webpage
soup = BeautifulSoup(response.content, 'html.parser')

# Find all image elements in the webpage
images = soup.find_all('img')

# Extract the image source URLs
img_urls = [img['src'] for img in images]

# Create a directory to save the images
if not os.path.exists('images'):
    os.makedirs('images')

# Download and save each image
for i, url in enumerate(img_urls):
    if url.startswith("http"):
        response = requests.get(url)
        file_name = f"images/image_{i}.jpg"
        with open(file_name, "wb") as f:
            f.write(response.content)
        print(f"Image {i+1} downloaded: {file_name}")
```

```python
import requests
from bs4 import BeautifulSoup
import os

# URL of the Wikipedia page
url = 'https://en.wikipedia.org/wiki/List_of_best-selling_automobiles'

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Find all the image tags
img_tags = soup.find_all('img')

# Create a directory to save the images
if not os.path.exists('images'):
    os.makedirs('images')

# Download and save the images
for img in img_tags:
    src = img['src']
    if src.startswith('//'):
        src = 'https:' + src
    elif src.startswith('/wiki/'):
        continue
    elif not src.startswith('http'):
        continue
    response = requests.get(src)
    filename = src.split('/')[-1]
    with open('images/' + filename, 'wb') as f:
        f.write(response.content)

# Display the downloaded images
import glob
import matplotlib.pyplot as plt

# Get a list of all the image files
image_files = glob.glob('images/*.jpg') + glob.glob('images/*.jpeg') + glob.glob('images/*.png')

# Display the images
for i, img_file in enumerate(image_files):
    plt.subplot(1, len(image_files), i+1)
    plt.imshow(plt.imread(img_file))
    plt.axis('off')

plt.show()
```
